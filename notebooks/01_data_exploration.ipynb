# 01 – Data Exploration

In this notebook we explore the Default of Credit Card Clients dataset.

```
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('../data/processed/credit_default_processed.csv')

# Display first five rows
df.head()
```

```
# Summary statistics
df.describe()
```

The dataset contains 30,000 observations and 25 columns with no missing values.

```
# Correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```
